```python
from pathlib import Path
import glob
# replace with path to folder of .txt files that contains all the ingredients as one string
text_files = glob.glob("ingredients" + "/*.txt")
file_names = [Path(text).stem for text in text_files]
long_string = ""
# gets all the info from all the files
for file_ in file_names:
    with open("ingredients/" + file_ + ".txt", 'r', encoding='utf-8') as f:
        long_string += f.read()
```

```python
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
import string

# Tokenize the text and remove punctuation
tokens = word_tokenize(long_string)
tokens = [word.lower() for word in tokens if word.isalpha()]

# Count the frequency of each word
word_freq = Counter(tokens)
```

```python
import numpy as np
import nltk
import matplotlib.pyplot as plt
import seaborn as sns

# Take the most common words for the heatmap
most_common_words = word_freq.most_common(20)  # Adjust the number as needed
labels, counts = zip(*most_common_words)

# Create a heatmap
plt.figure(figsize=(17, 3))
sns.heatmap([counts], annot=True, xticklabels=labels, fmt="d")
plt.title("Word Frequency Heatmap")
plt.show()
```
